```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("tested.csv")
print(data.head())
```

```python
data.info()
data.shape
```

```python
data.describe()
```

```python
data['Fare'].fillna(data["Fare"].mean())
```

```python
data['Cabin']
```

```python
data['Age'].fillna(data["Age"].median(), inplace = True)
```

```python
data['Cabin']
```

```python
data = data.drop(columns=["Cabin"])
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='Survived', data=data)
plt.title('Survival Count')
plt.show()
```

```python
#surival rate according to pclass
sns.countplot(x = 'Pclass', data= data, hue = 'Survived')
plt.title("Survival according to PClass")
```

```python
sns.boxplot(y = 'Age', x= 'Survived', data = data)
plt.title("Survived vs Age")
```

```python
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.histplot(data["Age"] , kde = True)
plt.title("Count of Age")
```

```python
sns.barplot(x= 'Survived', y = 'Fare', data = data)
plt.title("Fare vs Survived")
```
